```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters of the Lorenz system
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0
delta_t = 0.01  # Time step

# Simulation time parameters
t_max = 50
num_steps = int(t_max / delta_t)
time = np.linspace(0, t_max, num_steps)

# Initialize the state variables (x1, x2, x3)
x1 = np.zeros(num_steps)
x2 = np.zeros(num_steps)
x3 = np.zeros(num_steps)

# Initial conditions
x1[0], x2[0], x3[0] = 1.0, 1.0, 1.0  # Example initial conditions

# Initialize output y
y = np.zeros(num_steps)

# Iterate and update the state variables using the discrete-time Lorenz equations
for t in range(1, num_steps):
    x1[t] = x1[t-1] + delta_t * sigma * (x2[t-1] - x1[t-1])
    x2[t] = x2[t-1] + delta_t * (x1[t-1] * (rho - x3[t-1]) - x2[t-1])
    x3[t] = x3[t-1] + delta_t * (x1[t-1] * x2[t-1] - beta * x3[t-1])

    # Observation equation: y(t) = x1(t) + 2 * x2(t)
    y[t] = x1[t] + 2 * x2[t]

# Create the horizontal subplots
fig = plt.figure(figsize=(14, 6))

# 3D plot of the state variables
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot(x1, x2, x3, lw=1.5)
ax1.set_title('3D Plot of State Variables (x1, x2, x3)')
ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
ax1.set_zlabel('x3')

# Plot the output y
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(time, y, label='y = x1 + 2 * x2', color='green')
ax2.set_title('Output y')
ax2.set_xlabel('Time')
ax2.set_ylabel('Output')
ax2.legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
